```python
censusDataSet_filename = os.path.join(os.getcwd(), "data", "censusData.csv")
df = pd.read_csv(censusDataSet_filename, header = 0)
```

Here I am loading the dqata agian to preapre again to allign with my plan

```python


X = df.drop('income_binary', axis=1) 
y = df['income_binary']
```

I am creating the appropriate lables here. I changed the orginal feature name so it can be more simple 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=1234)
```

```python
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
y_lr_pred = lr_model.predict(X_test)
```

Here I am creating my logitic regression by using the appropriate function. I also then fit the model and predict it. 

```python
rf_model = RandomForestClassifier(max_depth=32, n_estimators=300, random_state=1234)
rf_model.fit(X_train, y_train)
y_rf_pred = rf_model.predict(X_test)
```

Doing the same thing I did above but for randome forest

```python
lr_accuracy = accuracy_score(y_test, y_lr_pred)
lr_roc_auc = roc_auc_score(y_test, lr_model.predict_proba(X_test)[:, 1])
```

```python
rf_accuracy = accuracy_score(y_test, y_rf_pred)
rf_roc_auc = roc_auc_score(y_test, rf_model.predict_proba(X_test)[:, 1])

print('[LR] Accuracy: {0:.2f}'.format(lr_accuracy))
print('[LR] ROC-AUC: {0:.2f}'.format(lr_roc_auc))
print('[RF] Accuracy: {0:.2f}'.format(rf_accuracy))
print('[RF] ROC-AUC: {0:.2f}'.format(rf_roc_auc))
```

```python
models = ['Logistic Regression', 'Random Forest']
accuracies = [lr_accuracy, rf_accuracy]
roc_auc_scores = [lr_roc_auc, rf_roc_auc]
```

```python
x = np.arange(len(models))
width = 0.35 

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, accuracies, width, label='Accuracy')
bars2 = ax.bar(x + width/2, roc_auc_scores, width, label='ROC-AUC')

ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()


def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)

plt.ylim([0, 1])
plt.show()
```
